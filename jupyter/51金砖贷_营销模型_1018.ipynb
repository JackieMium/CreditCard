```python
# Y值预测
from sklearn.metrics import classification_report
from sklearn.metrics import recall_score, precision_score, accuracy_score

Y_rf_predict = base_tree.predict(x_rf_test)
print(classification_report(Y_rf_test, Y_rf_predict))
fpr,tpr,thresholds= roc_curve(np.array(Y_rf_test),Y_rf_predict)
print ('KS:',max(tpr-fpr))
print('recall: ', recall_score(Y_rf_test, Y_rf_predict))
```
